## Scraping with pandas

Pandas is an awesome library for dealing with data. Let's do some scraping with some tables we can find on wikipedia, and see if we transform it into a programmatic format we can use conveniently.

```python
## Again, you can use the BeautifulSoup library to find what one
url = "https://en.wikipedia.org/wiki/List_of_Texas_locations_by_per_capita_income"
page = request.urlopen(url)
soup = BeautifulSoup(page, "html.parser")


## In this case I want to try a specific table where I want to specify the attributes
## soup.find will find the FIRST instance of table where with wikitable
## if there's no such table, it will return None
## https://www.crummy.com/software/BeautifulSoup/bs4/doc/#find
html_table = soup.find('table', class_ = "wikitable")

## Now you can parse the html into dataframes using pd.read_html
dfs = pd.read_html(html_table.prettify(), header=0)

```
